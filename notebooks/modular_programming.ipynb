```python
class Calculator:
    def add(self, a, b):
        return a + b

    def divi(self, a, b):
        if b == 0:
            raise ValueError("Division by zero is not allowed.")
        return a / b


# Example usage:
calc = Calculator()
print(calc.add(5, 3))   # 8
print(calc.divi(10, 2)) # 5.0
```
